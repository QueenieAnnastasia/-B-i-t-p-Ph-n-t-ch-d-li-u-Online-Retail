### 1. Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính

```python
import pandas as pd
import numpy as np
df = pd.read_csv('OnlineRetail.csv', encoding = 'ISO-8859-1')
df
```

### 2. Tạo cột mới có tên quý - 'Precious' nhận giá trị 1 nếu ngày lập hoá đơn nằm trong các tháng 1,2,3; nhận giá trị 2 nếu ngày lập hoá đơn nằm trong các tháng 4,5,6; nhận giá trị 3 nếu ngày lập hoá đơn nằm trong các tháng 7,8,9; nhận giá trị 4 nếu ngày lập hoá đơn nằm trong các tháng 10,11,12;

```python
# lọc tháng
a = [x[3:4] for x in df['InvoiceDate']]
df['thang']= a
df.head()
```

```python
dk = [(df['thang']=='1')|(df['thang']=='2')|(df['thang']=='3'),(df['thang']=='4')|(df['thang']=='5')|(df['thang']=='6'),(df['thang']=='7')|(df['thang']=='8')|(df['thang']=='9'),(df['thang']=='10')|(df['thang']=='11')|(df['thang']=='12')]
quy = ['1','2','3','4']
df['Precious'] = np.select(dk,quy)
df
```

### 3. Tạo một cột mới có tên 'Amount' có giá trị bằng Quantity * UnitPrice

```python
df['Amount'] = df['Quantity']*df['UnitPrice']
df
```

### 4. Tạo một cột mới 'Discount' nhận giá trị 10% nếu Country là 'United Kingdom' và thuộc quý 4, 5% nếu là 'France' ngược lại là 0%

```python
dk2 = [(df['Country']== 'United Kingdom') & (df['Precious']== '4'),df['Country']=='France']
giamgia = [10,5]
df['Discount'] = np.select(dk2,giamgia, default = 0)
df
```

### 5. Tạo cột mới 'Total' nhận giá trị bằng: Amount - Discount

```python
df['Total'] = df['Amount']*(1-df['Discount']/100)
df
```
